```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
df = sm.datasets.macrodata.load_pandas().data
index = pd.Index(sm.tsa.datetools.dates_from_range('1959Q1', '2009Q3'))
df.index=index
```

```python
gdp_cycle,gdp_trend = sm.tsa.filters.hpfilter(df['realgdp'])
df['trend']=gdp_trend
df[['trend', 'realgdp']]["2000-03-01":].plot(figsize=(14,6))
```

```python
import quandl
df_2 = quandl.get('BITFINEX/ETHUSD',start_date='2016-01-01',end_date='2023-01-31', index_col='Date', parse_dates=True)
eth_cycle,eth_trend = sm.tsa.filters.hpfilter(df_2['Mid'])
df_2["Trend"] = eth_trend
df_2[['Mid', 'Trend']].plot(figsize=(14,6))
```

```python
#zoom-in the recent bullrun
df_2[['Mid', 'Trend']]['2020-06-01':].plot(figsize=(14,6))
```
