```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv(r'C:\Users\sereb\Downloads\chains (1).csv', index_col= "Date", parse_dates= True)
df = df[df.index >= '2021-09-14']
df.fillna(0, inplace=True)
```

```python
df['Rollups']= df['Optimism'] + df['Arbitrum']
df['SoLunAvax'] = df['Solana'] + df['Avalanche'] + df['Terra Classic'] + df['Terra2']
df = df[['Rollups', 'SoLunAvax']]
df
```

```python
df.plot(title='Rollups and SoLunAvax TVL', figsize=(16,6))
ax = plt.gca()
formatter = ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x/1e9) + 'bln')
ax.yaxis.set_major_formatter(formatter)
plt.show()
```

```python
df['Ratio']=df['SoLunAvax']/df['Rollups']
df['Ratio'].plot(title='Ratio of SoLunAvax to Arbitrum+Optimism TVL', figsize=(16,6))
```
